```python
WES_muts = pd.read_csv('/Users/lincoln.harris/Desktop/WES_mutations_run.csv')
WES_muts
```

```python
found_muts = list(WES_muts.columns)
found_muts.remove('sample')

syno_muts = []
non_syno_muts = []

for x in found_muts:
    mut = x.split('_')[1]
    if mut[0] == mut[-1]:
        syno_muts.append(x)
    else:
        non_syno_muts.append(x)
```

```python
len(non_syno_muts)
```

```python
non_syno_muts.insert(0, 'sample') # add sample col back in
WES_muts_trimmed = WES_muts[non_syno_muts]
WES_muts_trimmed
```

```python
# init empty dict
patients_dict = {}

for idx, row in WES_muts_trimmed.iterrows():
    curr_sample = row['sample']
    curr_sample = curr_sample.split('_')[0]
    
    patients_dict.update({curr_sample : []})
```

```python
# fill it in 
for idx, row in WES_muts_trimmed.iterrows():
    curr_sample = row['sample']
    curr_sample = curr_sample.split('_')[0]
    
    curr_val = samples_dict[curr_sample]
    
    for i in range(1, len(row)): # starting at 1 to avoid sample name
        curr_ratio = row[i]
        curr_mut = row.index[i]
        
        if curr_ratio != '0:0':
            curr_val.append(curr_mut)
    
    patients_dict.update({curr_sample : curr_val})
```
