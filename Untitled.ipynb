```python
cells_df = pd.read_table(f, delim_whitespace=True, header=None, names=['is_prefix', 'cell_name'])
cells_df
```

```python
cells_df['fullPath'] = 's3://darmanis-group/singlecell_lungadeno/non_immune/nonImmune_fastqs_9.27/171120_concat/' + cells_df['cell_name']
cells_df
```

```python
cells_df['cell_name'] = cells_df['cell_name'].str.strip('/')
cells_df
```

```python
def get_fastqs_R1(cell):
    s3_location = f'{bucketPrefix}{cell}'
    lines = ! aws s3 ls $s3_location
    try:
        fq_line = [x for x in lines if x.endswith(('R1_001_merged.fastq.gz','R1_001.fastq.gz'))][0] # get the R1 fastq files
        fq_basename = fq_line.split()[-1]
        return f'{s3_location}{fq_basename}'
    except IndexError:
        return
```

```python
cells_df['input_fq_1'] = cells_df['cell_name'].map(get_fastqs_R1) 
cells_df
```
