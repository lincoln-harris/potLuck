```python
def search_for_sample_matches(df):
    """ """
    for sample in df.columns:
        keep = meta.sample_name == sample
        meta_sub = meta[keep]
        meta_sub = meta_sub.reset_index(drop=True)
        sample_mut_str = meta_sub.driver_gene.iloc[0] + ' ' + meta_sub.driver_mutation.iloc[0]

        muts_found = list(df[sample])

        if 'EGFR K745_A750T' in muts_found:
            muts_found.append('EGFR del19')

        if sample_mut_str in muts_found:
            #print(sample_mut_str)
            print(sample)
```
