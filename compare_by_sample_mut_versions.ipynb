```python
def value_counts_dict(elm):
    """ return a dict of every element in df, and the
        number of times its found """
    
    if elm in d1:
        curr_k = d1.get(elm)
        curr_k += 1
        d1.update({elm:curr_k})
    else:
        d1.update({elm:1})
```

```python
global d0, d1
d0 = {}
d1 = {}

v0.applymap(value_counts_dict)
d0
```

```python
v1.applymap(value_counts_dict)
d1
```

```python
d0_clean = {k: d0[k] for k in d0 if not pd.isna(k)} # dict comprehension
d0_clean.pop('BRAF ', None) # remove this empty BRAF key
d0_clean
```
