```python
import pandas as pd
pd.options.display.max_columns = None
```

```python
meta = pd.read_csv('/Users/lincoln.harris/code/SNP_calling_pipeline/metadata_all_cells_4.10.19.csv')

tumor_cells_df = pd.read_csv('/Users/lincoln.harris/Desktop/LAUD_important_shit/tumor_cells_190429.csv')
tumor_cells_list = list(tumor_cells_df.x)
```

```python
coverage_dataframe_tumor_cells = pd.read_csv('/Users/lincoln.harris/code/SNP_calling_pipeline/coverage/coverage_ratios_tumor_cells.csv')
coverage_dataframe_tumor_cells
```

```python
# want to see if im getting ANY hits to LT_S82
#    working now! whoo hoo!! 
for idx, row in coverage_dataframe_tumor_cells.iterrows():
    currCell = row.cells
    
    if currCell in tumor_cells_list: # got a tumor cell   
        keepRow = meta.cell_id == currCell  # match to metadata
        metaMatch = meta[keepRow]
        sampleName = list(metaMatch.sample_name)[0]
        
        if sampleName == 'LT_S82': # match to sample of interest
    
            for i in range(0, len(row.index)):
                curr_mut = row.index[i]
                curr_val = row[i]
                
                if curr_val != '0:0' and curr_mut != 'cells':
                    print(curr_mut)
```

```python
# get a list of all the genes we find mutations in 
genesList = []

for elm in list(coverage_dataframe_tumor_cells.columns):
    gene = elm.split('_')[0]
    genesList.append(gene)
    
genesList.remove('cells')
genesList = list(set(genesList))
genesList
```

```python
# get list of samples in metadata
samplesList = []
for idx, row in coverage_dataframe_tumor_cells.iterrows():
    currCell = row.cells
    
    if currCell in tumor_cells_list: # got a tumor cell   
        keepRow = meta.cell_id == currCell  # match to metadata
        metaMatch = meta[keepRow]
        sampleName = list(metaMatch.sample_name)[0]
        
        samplesList.append(sampleName)
        
samplesList = list(set(samplesList))
samplesList
```

```python
total_counts = pd.DataFrame(columns=samplesList, index=genesList)
total_counts[:] = 0 # set everybody to zero

variant_counts = pd.DataFrame(columns=samplesList, index=genesList)
variant_counts[:] = 0 # set everybody to zero
```

```python
# fill in total_counts df
for idx, row in coverage_dataframe_tumor_cells.iterrows():
    currCell = row.cells
    
    if currCell in tumor_cells_list: # got a tumor cell   
        keepRow = meta.cell_id == currCell  # match to metadata
        metaMatch = meta[keepRow]
        sampleName = list(metaMatch.sample_name)[0]
        
        for i in range(0, len(row.index)):
            curr_mut = row.index[i]
            curr_val = row[i]
            
            curr_mut_gene = curr_mut.split('_')[0]
            
            if curr_val != '0:0' and curr_mut != 'cells':
                v_count = int(curr_val.split(':')[0])
                t_count = int(curr_val.split(':')[1])
                
                total_counts[sampleName][curr_mut_gene] += t_count
```

```python
# fill in variant_counts df
for idx, row in coverage_dataframe_tumor_cells.iterrows():
    currCell = row.cells
    
    if currCell in tumor_cells_list: # got a tumor cell   
        keepRow = meta.cell_id == currCell  # match to metadata
        metaMatch = meta[keepRow]
        sampleName = list(metaMatch.sample_name)[0]
        
        for i in range(0, len(row.index)):
            curr_mut = row.index[i]
            curr_val = row[i]
            
            curr_mut_gene = curr_mut.split('_')[0]
            
            if curr_val != '0:0' and curr_mut != 'cells':
                v_count = int(curr_val.split(':')[0])
                t_count = int(curr_val.split(':')[1])
                
                variant_counts[sampleName][curr_mut_gene] += v_count
```
