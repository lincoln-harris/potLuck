```python
import os
import pandas as pd
pd.options.display.max_colwidth = 500
```

```python
bucketPrefix = 's3://darmanis-group/singlecell_lungadeno/rawdata/bam/181120/'
f = 'prefixe.txt'
! aws s3 ls $bucketPrefix > $f
! cat $f
```

```python
big_df = pd.read_table(f, delim_whitespace=True, header=None, names=['date', 'time', 'size', 'file'])
big_df
```

```python
# add a full_path col
big_df['full_path'] = 's3://darmanis-group/singlecell_lungadeno/rawdata/bam/181120/' + big_df['file']
big_df
```

```python
# add a cell_name col -- brilliant
big_df['cell_name'] = big_df['file'].str.split('_', expand=True)[0] + '_' + big_df['file'].str.split('_', expand=True)[1]
big_df
```

```python
# get rid of all of the files we dont want
sub_df = big_df[~big_df.file.str.contains(".tgz")]
sub_df = sub_df[~sub_df.file.str.contains(".tab")]
sub_df = sub_df[~sub_df.file.str.contains(".txt")]
sub_df = sub_df[~sub_df.file.str.contains("log.final.out")]
sub_df
```

```python
# make new s3 folders
dest = 's3://darmanis-group/singlecell_lungadeno/non_immune/nonImmune_bams_9.27/'
cmdList = []

for full_path in sub_df.full_path:
    file = full_path.split('/')[7]
    cellName = file.split('_')[0] +'_' + file.split('_')[1]
    cmd = 'aws s3 cp' + ' ' + full_path + ' ' + dest + '181120/' + cellName + '/' + file
    cmdList.append(cmd)
```

```python
cmdList
```

```python
with open('cmdList1.txt', 'w') as f:
    for item in cmdList:
        f.write("%s\n" % item)
```
