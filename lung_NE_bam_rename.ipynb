```python
old_bams_df = old_bams_df.rename(columns={'cell_name':'cell_name_old'})
old_bams_df['cell_name_new'] = ''
```

```python
# get reassigned cell_IDs
for idx, row in old_bams_df.iterrows():
    curr_cell = row.cell_name_old
    new_cell_name = ''
    
    rename_df_row = cell_rename_df[cell_rename_df.initial_name == curr_cell]
    try:
        new_cell_name = list(rename_df_row['actual_name'])[0]
        row.cell_name_new = new_cell_name
    except IndexError:
        row.cell_name_new = float('NaN')
```

```python
old_bams_df['file_name_new'] = ''
old_bams_df['s3_path_new'] = ''
```

```python
for idx, row in old_bams_df.iterrows():
    curr_file = row.file_name
    curr_cell_name_new = row.cell_name_new
    curr_file_name_new = ''
    
    try:
        if '.bai' in curr_file:
            curr_file_name_new = curr_cell_name_new + '.bam.bai'
        else:
            curr_file_name_new = curr_cell_name_new + '.bam'
        
        row['file_name_new'] = curr_file_name_new
    except TypeError:
        print(row)
```
