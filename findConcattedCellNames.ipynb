```python
concattedFilesList = pd.read_table(f, delim_whitespace=True, header=None, names = ['date','time', 'size', 'fileName'])
concattedFilesList
```

```python
concattedFilesList['fileName'].str.split('-')
```

```python
concattedFilesList['fileName'].str.split('-')[100][0]
concattedFilesList['fileName'].str.split('-')[100][1]
```

```python
concattedCellsList = []

for i in range(0, len(concattedFilesList.index)):
    curr = str(concattedFilesList['fileName'][i])
    currSplit = curr.split('-')
    try:
        addStr = currSplit[0] + '_' + currSplit[1]
        concattedCellsList.append(addStr)
    except IndexError:
        pass
```

```python
concattedCellsList
```

```python
concattedCells_df = pd.DataFrame(concattedCellsList)
concattedCells_df
```

```python
concattedCells_df.columns = ['cellName']
concattedCells_df
```

```python
concattedCells_df['run1'] = ''
concattedCells_df['run2'] = ''
concattedCells_df['run3'] = ''
concattedCells_df
```

```python
for line in concattedCells_df['cellName']:
    currPlate = line.split('_')[1]
    metaRow = metadata[metadata['plate'].isin([currPlate])] # find corresponding plate in metadata table
    run1 = metaRow['Sequence_Run1']
    run2 = metaRow['Sequence_Run2']
    run3 = metaRow['Sequence_Run3']
    run1 = str(run1).split()[1].strip('/')
    run2 = str(run2).split()[1].strip('/')
    run3 = str(run3).split()[1].strip('/')
    
    concattedCells_df['run1'] = run1
    concattedCells_df['run2'] = run2
    concattedCells_df['run3'] = run3
```
