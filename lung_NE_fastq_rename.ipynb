```python
import pandas as pd
```

```python
cell_rename_df = pd.read_csv('/Users/lincoln.harris/Desktop/cell_rename_assignments_lung_NE.csv')
cell_rename_df
```

```python
initial_name_redux = [x.split('-')[0] + '_' + x.split('-')[1] 
                      for x in cell_rename_df.initial_name]
```

```python
actual_name_redux = [x.split('-')[0] + '_' + x.split('-')[1] 
                      for x in cell_rename_df.actual_name]
```

```python
cell_rename_df.initial_name = initial_name_redux
cell_rename_df.actual_name = actual_name_redux

cell_rename_df
```

```python
concat_path = 's3://darmanis-group/singlecell_lungadeno/rohan_analysis/singlecell_lungne/concatenated/'

! aws s3 ls $concat_path > out.txt
! cat out.txt
```

```python
# read this bugger in 
concat_s3_df = pd.read_csv('out.txt', names=['date', 'time', 'size', 'filename'], delim_whitespace=True)
concat_s3_df = concat_s3_df.drop(0)
concat_s3_df = concat_s3_df.reset_index(drop=True)

# dont really care about these other fields
concat_s3_df = concat_s3_df.drop(['date', 'time', 'size'], axis=1)
concat_s3_df
```

```python
concat_s3_df['s3_path_old'] = concat_path + concat_s3_df.filename
```

```python
concat_s3_df
```

```python
cellnames = [x.split('-')[0] + '_' + x.split('-')[1] for x in concat_s3_df.filename]
cellnames = [x.split('_')[0] + '_' + x.split('_')[1] for x in cellnames]

concat_s3_df['cell_name_orig'] = cellnames
concat_s3_df['cell_name_new'] = ''
concat_s3_df
```

```python
# get reassigned cell_IDs
for idx, row in concat_s3_df.iterrows():
    curr_cell = row.cell_name_orig
    new_cell_name = ''
    
    rename_df_row = cell_rename_df[cell_rename_df.initial_name == curr_cell]

    new_cell_name = list(rename_df_row['actual_name'])[0]
    row.cell_name_new = new_cell_name
```

```python
concat_s3_df
```

```python
concat_s3_df['filename_new'] = ''

for idx, row in concat_s3_df.iterrows():
    curr_filename = row.filename
    new_filename = row.cell_name_new
    
    if 'R1' in curr_filename:
        new_filename = new_filename + '_R1.fastq.gz'
    elif 'R2' in curr_filename:
        new_filename = new_filename + '_R2.fastq.gz'

    row.filename_new = new_filename
```

```python
# not quite yet
s3_path_new = 's3://darmanis-group/singlecell_NE/fastq_rename_11.5.19/'
concat_s3_df['s3_path_new'] = s3_path_new + concat_s3_df.filename_new
```

```python
s3_cmds = 'aws s3 cp ' + concat_s3_df.s3_path_old + ' ' + concat_s3_df.s3_path_new
```

```python
concat_s3_df['s3_cmds'] = s3_cmds
concat_s3_df
```

```python
concat_s3_df.s3_cmds.to_csv('s3_cmds_fastq.csv', index=False)
```
