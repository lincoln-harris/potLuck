```python
import pandas as pd
```

```python
val = pd.read_csv('/Users/lincoln.harris/Desktop/validationTable_cells_fathmm_filter.5.22.19.csv', index_col=0)
val
```

```python
muts_found_list = list(val.mutations_found)
muts_found_list
```

```python
muts_dict = {}
for entry in muts_found_list:
    
    if not str(entry) == 'nan':
        entry_split = entry.split(',')

        for sub_entry in entry_split:
            sub_entry = sub_entry.rstrip()
            sub_entry = sub_entry.lstrip()
            
            try:
                aa_str = sub_entry.split(' ')[1]
                if aa_str[0] != aa_str[-1]: # make sure its a non-synon mut
                    if sub_entry in muts_dict.keys():
                       muts_dict[sub_entry] += 1
                    else:
                       muts_dict[sub_entry] = 1
            except IndexError: # just an empty string
                print(entry)
```

```python
muts_dict
```

```python
# sort?? 
import operator

muts_dict = sorted(muts_dict.items(), key=operator.itemgetter(1))
muts_dict
```

```python
# these are the most common MUTATIONS
list(reversed(muts_dict))
```

```python
# now lets find the most common CO-MUTATIONS
from itertools import combinations

comuts_dict = {}
for entry in muts_found_list:
    
    if not str(entry) == 'nan':
        entry = entry.rstrip()
        entry = entry.lstrip()
        num_muts = len(entry.split(','))
        
        if num_muts == 2: # easy case
            if entry in comuts_dict.keys():
                comuts_dict[entry] += 1
            else:
                comuts_dict[entry] = 1
                
        elif num_muts > 2: # hard case -- need to take combinations
            combos = combinations(entry.split(','), 2)
            for combo in combos:
                new_entry = combo[0].rstrip().lstrip() + ', ' + combo[1].rstrip().lstrip()
                print(new_entry)
                if new_entry in comuts_dict.keys():
                    comuts_dict[new_entry] += 1
                else:
                    comuts_dict[new_entry] = 1
```

```python
# remove syno-muts -- think this works!!
comuts_dict_clean = comuts_dict.copy()

for key in comuts_dict.keys():
    #print(key)
    for elm in key.split(','):
        elm_c = elm.rstrip().lstrip()
        
        try:
            aa = elm_c.split(' ')[1]
            if aa[0] == aa[-1]: # got a syno-mut, need to pop
                del comuts_dict_clean[key]
                break
        except IndexError: # again just a blank entry 
            print(key)
```

```python
#sort??
comuts_dict_clean_sort = sorted(comuts_dict_clean.items(), key=operator.itemgetter(1))
comuts_dict_clean_sort
```

```python
comuts_df = pd.DataFrame(comuts_dict_clean_sort, columns=['pair', 'count'])
comuts_df
```
