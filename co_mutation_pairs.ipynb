```python
def remove_syno_muts():
    ''' removes synonomous mutations from dict '''
    clean = comuts_dict.copy()

    for key in comuts_dict.keys():
        for elm in key.split(','):
            elm_c = elm.rstrip().lstrip()

            try:
                aa = elm_c.split(' ')[1]
                if aa[0] == aa[-1]: # got a syno-mut, need to pop
                    del clean[key]
                    break   # break inner loop
            except IndexError: # again just a blank entry 
                print(key)
                
    return(clean)
```

```python
def update_dict(m, g):
    ''' element already in dict '''
    if g == 'naive':
        comuts_dict[m][0] += 1
    elif g == 'grouped_pr':
        comuts_dict[m][1] += 1
    elif g == 'grouped_pd':
        comuts_dict[m][2] += 1
```

```python
def add_to_dict(m, g):
    ''' element not in dict '''
    if g == 'naive':
        comuts_dict[m] = [1,0,0]
    elif g == 'grouped_pr':
        comuts_dict[m] = [0,1,0]
    elif g == 'grouped_pd':
        comuts_dict[m] = [0,0,1]
```

```python
def process_mutation(mut_str, group):
    ''' encapsulating functionality -- add to comuts_dict '''
    
    mut_str = mut_str.rstrip().lstrip()
    num_muts = len(mut_str.split(','))
    
    if num_muts == 2:   # easy case
        if mut_str in comuts_dict.keys():
            update_dict(mut_str, group)
        else:
            add_to_dict(mut_str, group)
    
    elif num_muts > 2:   # hard case -- need to take combinations
        combos = combinations(mut_str.split(','), 2)
        
        for combo in combos:
            new_mut_str = combo[0].rstrip().lstrip() + ', ' + combo[1].rstrip().lstrip()
            if new_mut_str in comuts_dict.keys():
                update_dict(new_mut_str, group)
            else:
                add_to_dict(new_mut_str, group)
```

```python
# MAIN logic and driver loop 
global comuts_dict
comuts_dict = {}
    # [0, 1, 2]  
            # 0 - naive
            # 1 - grouped_pr
            # 2 - grouped_pd

for cell in list(val_tbl.index): 
    muts = val_tbl.loc[cell].mutations_found
    sample = val_tbl.loc[cell].sample_name
    
    if not str(sample) == 'nan':
        meta_line = meta[meta.sample_name == sample].iloc[0] # will return a df with every cell for that sample;
        group = meta_line.analysis                          # i just care about the `group`, so taking the 0th line 
        
        if not str(muts) == 'nan': # valid mut str
            process_mutation(muts, group)
        

comuts_dict_clean = remove_syno_muts()
comuts_dict_clean   # nice!!
```
