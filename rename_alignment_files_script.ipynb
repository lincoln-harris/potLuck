```python
import pandas as pd
pd.set_option('display.max_rows', 999)
```

```python
# read s3 output to csv
out = 'out.csv'
PATH = 's3://darmanis-group/singlecell_NE/alignment_7.25.19/'
! aws s3 ls $PATH > $out
```

```python
# get just the file names
s3_files_df = pd.read_csv('out.csv', names=['date', 'time', 'size', 'name'], delim_whitespace=True)
s3_files_names = s3_files_df.name
s3_files_names
```

```python
# want to get rid of first row
s3_files_names = s3_files_names.dropna()
s3_files_names
```

```python
# remove this weird internal string
new_names = list(map(lambda x: x.replace('Aligned.sortedByCoord.out', ''), s3_files_names))
new_names
```

```python
# replace dash with underscore
new_names = list(map(lambda x: x.replace('-', '_'), new_names))
new_names
```

```python
# init new df to keep track of everything
df = pd.DataFrame(s3_files_names)
df.columns = ['curr_file_name']
df
```

```python
# add in new file names
df['new_file_name'] = new_names
df
```

```python
# add in curr / new path cols!
#    holy shit that was easy
df['curr_path'] = PATH + df.curr_file_name
df['new_path'] = PATH + df.new_file_name
df
```

```python
# validate
print(df.curr_path[2501])
print(df.new_path[2501])
```

```python
# now add in the s3 mv cmds
cmd_line = 'aws s3 mv '
df['s3_cmd'] = cmd_line + df.curr_path + ' ' + df.new_path
df
```

```python
# validate -- perfect
df.s3_cmd[1191]
```

```python
df.s3_cmd.to_csv('alignment_s3_mv_cmds.csv', index=False, header=None)
```
