```python
import os
import pandas as pd
pd.options.display.max_colwidth = 500 # module config? 
```

```python
# get all of the run prefixes w/in immuneCells_9.27
bucketPrefixes = 's3://darmanis-group/singlecell_lungadeno/rawdata/fastqs/'
f = 'prefixes.txt'
! aws s3 ls $bucketPrefixes > $f
! cat $f
```

```python
runs_df = pd.read_table(f, delim_whitespace=True, header=None, names=['is_prefix', 'run_name'])
runs_df
```

```python
# can i add a full_path col? 
runs_df['full_path'] = 's3://darmanis-group/singlecell_lungadeno/rawdata/fastqs/' + runs_df['run_name']
runs_df
```

```python
runs_df = runs_df.drop([23])
runs_df
```

```python
# think this is working!

outDir = 's3://darmanis-group/singlecell_lungadeno/non_immune/nonImmune_fastqs_9.27/'
cmdList = []

for run in runs_df['full_path']:
    f1 = 'cellsOut.txt'
    ! aws s3 ls $run > $f1
    currCells_df = pd.read_table(f1, delim_whitespace=True, header=None, names=['date', 'time', 'size', 'file'])
    
    currRun = run.split('s3://darmanis-group/singlecell_lungadeno/rawdata/fastqs/')[1]
    currRun = currRun.split('_')[0]
    print(currRun)
    
    for cell in currCells_df['file']:
        try:
            cell_split = cell.split('_')
            currCell = cell_split[0] + '_' + cell_split[1]
            cmd = 'aws s3 cp ' + run + cell + ' ' + outDir + currRun + '/' + currCell + '/'
            cmdList.append(cmd)
            #! $cmd
        except AttributeError: # dont know why im seeing this error
            continue
```

```python
cmdList
```

```python
with open('cmdList.txt', 'w') as f:
    for item in cmdList:
        f.write("%s\n" % item)
```
